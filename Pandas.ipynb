# Filter

```python
# loc: filter rows by LABEL, and select columns by LABEL
hr.loc[1]                        # row with label 1
```

```python
hr.loc[1:3]                      # rows with labels 1 through 3
```

```python
hr.loc[1:3, 'Age':'MonthlyIncome']  # rows 1-3, columns 'age' through 'occupation'
```

```python
hr.loc[:, 'Age':'MonthlyIncome']    # all rows, columns 'age' through 'occupation'
```

```python
hr.loc[[1,3],['Age','MonthlyIncome']]  # rows 1 and 3, columns 'age' and 'gender'
```

```python
# iloc: filter rows by POSITION, and select columns by POSITION
hr.iloc[0]  
```

```python
hr.iloc[0:3]                     # rows with positions 0 through 2 (not 3)
```

```python
hr.iloc[0:3, 0:3]                # rows and columns with positions 0 through 2
```

```python
hr.iloc[:, 0:3]                  # all rows, columns with positions 0 through 2
```

```python
hr.iloc[[0,2], [0,1]]            # 1st and 3rd row, 1st and 2nd column
```

```python
# mixing: select columns by LABEL, then filter rows by POSITION

hr.Age[0:3]
```

```python
hr[['Age', 'Gender', 'Department']][0:3]
```
